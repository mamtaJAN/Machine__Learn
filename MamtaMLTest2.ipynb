```python
#iloc[]==>
#Used for integerlocation based indexing. we find the rows and columns by their integer positions.
import pandas as pd

# Create a sample DataFrame
data = pd.DataFrame({
    'Name': ['Anil', 'Bunty', 'Charlie', 'Dev'],
    'Age': [25, 30, 35, 40],
}, index=['a', 'b', 'c', 'd'])

print(data)
```

```python
value = data.loc['b', 'Age']
print(value) 

```

```python
value = data.iloc[0, 1]
print(value)  # Output: 25
```

```python
#dropna and fillna()
#dropna()==Used to remove missing values  from the DataFrame or Series.
#fillna()==Used to fill missing values  with a specified value or method.
data = {'A': [1, 2, 3, None], 'B': [4, None, 6, 7]}
df = pd.DataFrame(data)
data
print(df)
```
